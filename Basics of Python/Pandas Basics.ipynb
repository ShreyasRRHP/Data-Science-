### Merge function 

```python
#to combine two dataframes 
df1 = pd.DataFrame({'Roll no':[1,2,3,4,5],'Physics':[67,78,56,98,67]})
print(df1)
df2 = pd.DataFrame({'Roll no':[1,2,3,4,5],'Chemistry':[45,68,96,36,78]})
print(df2)
```

```python
#on function is on what basis to merge 
pd.merge(df1,df2, on = 'Roll no')
```

```python
#if not specific, then it will be auto to find the common  
pd.merge(df1,df2)
```

```python
df11 = pd.DataFrame({'Roll no':[1,2,3,4,5,8,9],'Physics':[67,78,56,98,67,68,89]})
print(df1)
df22 = pd.DataFrame({'Roll no':[1,2,3,4,5,7,6],'Chemistry':[45,68,96,36,78,34,56]})
print(df2)
```

```python
#if some of the commom is not present then it will reget it and get the ones which are common 
pd.merge(df11 , df22)
```

```python
#for dismilarity and we must put emphasis on perticular format 
pd.merge(df11 , df22 , how = 'left' ) 
#if it's non existane then the value is taken to zero  
#left means to put emphasis on the data frame on the left  
```

```python
pd.merge(df11 , df22 , how = 'right' ) 
#if it's non existane then the value is taken to zero  
#left means to put emphasis on the data frame on the right 
```

```python
pd.merge(df11 , df22 , how = 'outer' ) 
#if it's non existane then the value is taken to zero  
#and a new row are created for each of the unique
```

### pivot table 

```python
# as name suggests pivots the table according to the presented variables 
marks2 = pd.read_csv(r"E:\Coding\Data Science\EDA\dataset\collagemarks.csv")
marks2.head()
```

```python
pd.pivot_table(marks2, index='Branch', aggfunc='sum')
```

```python
marks2.dtypes
```

```python
# Convert columns to numeric, coercing errors to NaN
marks2['Maths'] = pd.to_numeric(marks2['Maths'], errors='coerce')
marks2['Physics'] = pd.to_numeric(marks2['Physics'], errors='coerce')
marks2['Chemistry'] = pd.to_numeric(marks2['Chemistry'], errors='coerce')

# Create pivot table specifying the numeric columns explicitly in `values`
pd.pivot_table(marks2, index='Branch', values=['Maths', 'Physics', 'Chemistry'], aggfunc='mean')
```

```python
pd.pivot_table(marks2, index='Branch', aggfunc='count')
```

```python
pd.pivot_table(marks2, index='Branch', aggfunc='max')
```

```python
pd.pivot_table(marks2, index='Branch', values=['Maths', 'Physics', 'Chemistry'], columns = 'Section',aggfunc='mean')
```
