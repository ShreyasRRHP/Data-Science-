```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load the example 'tips' dataset
tips = sns.load_dataset("tips")

# ---------------------------------------------
# 1. Line plot with default confidence interval
# ---------------------------------------------
# Seaborn automatically adds a translucent shaded area around the line.
# This shaded region is the 95% confidence interval (CI) or ((ci) in code) by default.
sns.lineplot(data=tips, x="size", y="tip")
plt.title("Tip vs Table Size with 95% Confidence Interval")
plt.show()

# ---------------------------------------------
# 2. Line plot with NO confidence interval
# ---------------------------------------------
# The 'ci' parameter is deprecated. Use 'errorbar=None' instead.
# This removes the translucent band entirely.
sns.lineplot(data=tips, x="size", y="tip", errorbar=None)
plt.title("Tip vs Table Size WITHOUT Confidence Interval")
plt.show()

# ---------------------------------------------
# 3. Line plot with custom (68%) confidence interval
# ---------------------------------------------
# You can specify a different CI level using: errorbar=('ci', level)
sns.lineplot(data=tips, x="size", y="tip", errorbar=('ci', 68))
plt.title("Tip vs Table Size with 68% Confidence Interval")
plt.show()
```

## hist Plot 

```python
# hist plot is just the histogram with kernel density estimate, with the definition being.
#A Kernel Density Estimate (KDE) is a way to estimate the probability density function (PDF) of a random variable based on a finite data sample. 
#In simple terms, it's a smoothed version of a histogram that helps visualize the distribution of data
```

```python
#fron the hr data 
df= pd.read_csv(r"E:\Coding\Data Science\EDA\dataset\hr_data.csv")
df.head()
```

```python
sa.histplot(df['time_spend_company'],kde = True)
```

```python
sa.histplot(df['average_montly_hours'],kde = True)
```

```python
# we can difine the range for the histogram with the help of bins function 
bins = [2,3,4,5,6,7,8,9,10]
sa.histplot(df['time_spend_company'],kde = True, bins = bins)
plt.xticks(bins);
```

```python
# to remove the kde line we can use the flase function 
bins = [2,3,4,5,6,7,8,9,10]
sa.histplot(df['time_spend_company'],kde = False, bins = bins)
plt.xticks(bins);
```

```python
#to remove the kde line we can use the following function 
bins = [2,3,4,5,6,7,8,9,10]
sa.kdeplot(df['time_spend_company'])
plt.xticks(bins);
```

```python
# to remove the kde line we can use the flase function 
bins = [2,3,4,5,6,7,8,9,10]
sa.histplot(df['time_spend_company'],
            kde = True,
            bins = bins,
           color='red',            # Histogram fill color
            edgecolor='blue',       # Outline color
            linewidth=3,
            alpha=0.5)
plt.xticks(bins);
```

```python
#to remove the kde line we can use the flase as kdg = false 
bins = [2,3,4,5,6,7,8,9,10]
plt.figure(figsize = (8,8))
sa.histplot(df['time_spend_company'],
            bins = bins,
           color='red',            # Histogram fill color
            edgecolor='blue',       # Outline color
            linewidth=3,
            alpha=0.5)

sa.kdeplot(
    df['time_spend_company'],
    color='green'
)
plt.xticks(bins);
```

## Heatmap

```python
fdf = sa.load_dataset('flights')
fdf.head()
```

```python
fdf = fdf.pivot(index="month", columns="year", values="passengers")  #  Correct
#beacuse the same space is being used hence pivoting again will cause errors 
fdf.head()
```

```python
plt.figure(figsize = (12,6))
ax = sa.heatmap(fdf)
ax
```

```python
# for the value to be incorporated we use the annot fucntion 
plt.figure(figsize = (12,6))
ax = sa.heatmap(fdf , annot = True , fmt = 'd')#fmt is format string value to make sure the values in decimal places 
ax
```

```python
#customizing with some other parameters 
plt.figure(figsize = (12,6))
ax = sa.heatmap(fdf , annot = True , fmt = 'd' , linewidth = '5' , linecolor = 'white' , cmap = 'Blues')
#fmt is format string value to make sure the values in decimal places 
ax
```

```python
#futher customization with 
grid_kws = {"height_ratios": (.4,.05),"hspace" : .4}
f , (ax , cbar_ax) = plt.subplots(2 , gridspec_kw = grid_kws)
ax = sa.heatmap(fdf , cbar_kws = {"orientation" : "horizontal"} , ax=ax ,cbar_ax = cbar_ax,)
```

```python
#heat map can also be used to study correlation between parameters as well 
#example=
tdf = sa.load_dataset('titanic')
tdf.head()
```

```python
# Compute the correlation matrix (only for numeric columns)
corr_matrix = tdf.corr(numeric_only=True)

# Display the correlation matrix
print(corr_matrix)
```

```python
sa.heatmap(corr_matrix)
```
