## Scatter plot 

```python
# for scatter plot let take a simple example 
Roll_no = np.arange(1,11)
Marks = np.arange(0,100,10)
print(Roll_no)
print(Marks)
```

```python
Class = pd.DataFrame({'roll_no':Roll_no,'marks':Marks})
print(Class)
```

```python
plt.scatter(Class['roll_no'],Class['marks'], color = "green" ,marker = 'v' , s = 150)
#the format being that plt.scatter(x-axis,y-axis,features....) feature include color size(S) and type of symbol of the representation 
#to lable both the axises 
plt.xlabel("Roll no")
plt.ylabel("Marks")
plt.title("the marks of class a")
plt.show()
```

```python
#changing the size of the image 
plt.figure(figsize = (12,8))
plt.scatter(Class['roll_no'],Class['marks'], color = "green" ,marker = 'v' , s = 150)
#the format being that plt.scatter(x-axis,y-axis,features....) feature include color size(S) and type of symbol of the representation 
#to lable both the axises 
plt.xlabel("Roll no")
plt.ylabel("Marks")
plt.title("the marks of class a")
plt.show()
```

```python
subjects = ['Maths', 'English', 'Science', 'Social Studies', 'Computer']
marks1 = [89, 90, 45, 78, 99]
marks2 = [78, 56, 34, 90, 12]
```

```python
plt.figure(figsize = (8,8))

plt.bar(subjects, marks1)
plt.bar(subjects, marks2)

plt.xlabel('Subjects')
plt.ylabel('Marks')

plt.show()
```

```python
#to make the plot side by side
#take the length of the subjects and try to allocate 40% of the space to one marks and the other 40 to the second marks, 
#with which we will be able to manintain 
#space 
Sub_len = np.arange(len(subjects))
width = 0.4
```

```python
plt.figure(figsize = (8,8))

plt.bar(Sub_len , marks1 , width = width)
plt.bar(Sub_len + width , marks2 , width = width)

plt.xlabel("subjects")
plt.ylabel("marks")
plt.show()
```

```python
#we can have a different appprouch with multiple colours and different transparency 
colors = ['red', 'blue', 'green', 'orange', 'purple']
plt.figure(figsize = (8,8))

plt.bar(Sub_len , marks1 , width = width , color = colors)
plt.bar(Sub_len + width , marks2 , width = width, color = colors , alpha = 0.5)

plt.xlabel("subjects")
plt.ylabel("marks")
plt.show()
```

## Pie Plot 

```python
classes = ['physics','chemistry','maths','sst','biology']
marks = [98,78,79,90,99]

colors = ['red', 'blue', 'green', '#9803fc', '#03c2fc']
```

```python
plt.pie(marks,labels = classes, colors = colors , autopct = '%0.1f%%')#autopct shows the percentage of the weights 
plt.show()
```

```python
# A way to differentiate a particular segment is by explode_value function, where 
explode_values = [0.1, 0, 0, 0, 0]# a detail for which to exclude 
plt.pie(marks,labels = classes, colors = colors , autopct = '%0.1f%%' , explode = explode_values, shadow = True, radius = 1.6)
#autopct shows the percentage of the weights, and shadow gives a shadow effect and the radius to enlarge or shorten the image 
plt.show()
```

```python
textprops = {'fontsize':14, 'color':'purple'}
#customizing the text in the image and adding the border with specifications 
wedgeprops = {'linewidth':3 , 'linestyle':'--' , 'edgecolor' : 'purple'}
plt.pie(marks, labels = classes,
        colors=colors, 
        autopct = '%0.1f%%', 
        explode=explode_values, 
        radius = 1.6,
        textprops = textprops,
        wedgeprops  = wedgeprops
       )
plt.show()
```

```python
plt.figure(figsize = (6,6))

plt.pie(marks, labels = classes, 
        colors=colors, 
        autopct = '%0.1f%%', 
        explode=explode_values, 
        textprops = textprops)
#the percentage can also be considered as average and legend 
plt.title('Subjects and Average Scores')
plt.legend()
plt.show()
```

## Subplots and export of images 

```python
# to plot multiple graphs in one place is by creating a grid system 
#basic data 
study_hours = [2,3,4,4, 5, 6, 7, 7, 8, 9, 9, 10, 11, 11, 12]
marks = [6, 10, 15, 20, 34, 44, 55, 60, 55, 67, 70, 80, 90, 99, 100]
```

```python
#Subplot can only be done with plt.figure function and mainly subplot function 
# here the subplot function is used to define the number of sub graphs needed and the format (x,y,z) indicated (x,y) 
# the matrix shape of the subplot and z the location in the from 
#(1,2)
#(3,4)
#make sure have the show function after the savefig if not a blank image will appear 

plt.figure(figsize = (8,8))

plt.subplot(2,2,1)
plt.plot(study_hours , marks , 'r-')

plt.subplot(2,2,2)
plt.scatter(study_hours , marks , color = 'blue')

plt.subplot(2,2,3)
plt.hist(marks)

plt.subplot(2,2,4)
plt.pie(marks,labels = study_hours)

#we use the savefig function to save the image to the device 
plt.savefig(r'C:\Users\DELL\Desktop\images\subplot.png')

plt.show()
```

```python
#another form would be by 
#Subplot can only be done with plt.figure function and mainly subplot function 
plt.figure(figsize = (8,8))

plt.subplot(1,2,1)
plt.plot(study_hours , marks , 'r-')

plt.subplot(1,2,2)
plt.scatter(study_hours , marks , color = 'blue')

plt.subplot(2,2,3)
plt.hist(marks)

plt.savefig(r"C:\Users\DELL\Desktop\images\Secondimage.jpg")

plt.show()
```
