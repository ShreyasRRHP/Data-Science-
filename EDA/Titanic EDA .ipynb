## Data Cleaning 

From the above data we can say that a large portion of the cabin data is not present and with no other methods of obtaining this information we will drop the column, the info of the cabin id is not an important data while considering the possibilty of survival of the passengers, even if we were to consider the class as factor in determining the porbability of survival we have the pclass data which can give insights on the issue 
we will also drop the name as that data would be of much relevence in the contex of the given task which is to find insgights on the srvival of the people in the ship 
for similar reason we will be dropping the ticket as much of the information is presnet in the embarked column 

```python
df= df.drop(['Cabin', 'Ticket', 'Name'] , axis =1 )
df
```

With missing data in the age column we try to make up for this by using mean values of the data set for compensation a simple method is by filling in the mean of that perticular column as the g

```python
print(df['Age'].median())
df['Age'].fillna(df['Age'].median())
```

We can try a better way by tring to get specific data on the groups as we the median by gender, class and survival try to get more insights  

```python
print("we can habe all the vriation and thier median by the group by method")
median = df.groupby(by = ['Sex','Pclass','Survived'])['Age'].median()
print(median)
print("The Variance if of " , df.groupby(by = ['Sex','Pclass','Survived'])['Age'].median().var() , '\n')
print("and the range of the data" , median.max() - median.min())
```

With such variable range of data simply replacing the null values with median would not be the most accurate thing replace with just the median hence we have to try to fill the values with these specific values 

```python
df['Age'] = df.groupby(by = ['Sex','Pclass','Survived'])['Age'].transform(lambda x : x.fillna(x.median()))
df.head(10)
```
